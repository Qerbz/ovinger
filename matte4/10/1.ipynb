```python
import numpy as np

def w(n):
    return np.round(np.e**(-1j*2*np.pi/n),16)

def F(N):
    return np.array([np.array([w(N)**(n*k) for n in range(N)]) for k in range(N)])

def ftransform(f):
    return np.round(np.matmul(F(len(f)),f),14)

f_arr = np.array([3,2,-1,1,-1,2,0,3,1,-2])
ftransform(f_arr)
```

```python
def f(t):
    return np.tan(np.sin(np.pi*2*t))

N = 200
f_arr = f(np.linspace(0,2,200))

f_arr_transformed = ftransform(f_arr)

import matplotlib.pyplot as plt

plt.plot(np.abs(f_arr_transformed[:20])/N)

filter = lambda x: abs(x)/N >= 0.05
print(filter(f_arr_transformed))
print(f_arr_transformed)
f_arr_transformed_filtered = filter(f_arr_transformed) * f_arr_transformed
print(f_arr_transformed_filtered)
plt.plot(np.abs(f_arr_transformed_filtered[:20])/N)
plt.show()
```

2, 6, 10 and 14 are visible

```python
filtered_signal = np.matmul(np.conjugate(F(len(f_arr)))/N,f_arr_transformed_filtered)
plt.plot(filtered_signal)
plt.plot(f_arr)
```
