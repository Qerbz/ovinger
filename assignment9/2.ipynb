```python
from matplotlib import pyplot as plt
import numpy as np
a0 = np.pi/16
n=5
a = [a0] + [np.sin(n * np.pi / 2)/(2 * n) + (1/(n*n*np.pi)) * (np.cos(n*np.pi/2)-1) for n in range(1,n)]
b = [0] + [1/(n*n*np.pi)*np.sin(n*np.pi/2)-(1/(2*n)) * np.cos(n*np.pi/2) for n in range(1,n)]
def S(x):
    sum = a0
    for i in range(1,n):
        sum += a[i]*np.cos(i*x)+b[i]*np.sin(i*x)
    return sum

x= np.arange(-3*np.pi,3*np.pi,0.1)
plt.plot(x,S(x))

```
